```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Dataset

```python
sales_data = pd.read_excel('ECOMM DATA.xlsx')
sales_data
```

```python
sales_data.head()
```

## Finding Revenue

```python
sales_data['Total Revenue'] = sales_data['Sales']
total_sales = sales_data['Total Revenue'].sum()
print(f"Total Sales Revenue: ${total_sales:.2f}")
```

## Analyzing Sales Trends

```python
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])
sales_data['Order Month'] = sales_data['Order Date'].dt.to_period('M')
order_monthly_sales = sales_data.groupby('Order Month')['Total Revenue'].sum()
print(order_monthly_sales)
```

```python
sales_data['Ship Date'] = pd.to_datetime(sales_data['Ship Date'])
sales_data['Ship Month'] = sales_data['Ship Date'].dt.to_period('M')
ship_monthly_sales = sales_data.groupby('Ship Month')['Total Revenue'].sum()
print(ship_monthly_sales)
```

## Visualizing analyzed trend

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=order_monthly_sales.index.astype(str), y=order_monthly_sales.values, marker='o',label='Order Date')
plt.title('Monthly Sales Trends')
plt.xlabel('Order Month')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=ship_monthly_sales.index.astype(str), y=ship_monthly_sales.values, marker='o',label='Ship Date')
plt.title('Monthly Sales Trends')
plt.xlabel('Ship Month')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=order_monthly_sales.index.astype(str), y=order_monthly_sales.values, marker='o',label='Order Date')
sns.lineplot(x=ship_monthly_sales.index.astype(str), y=ship_monthly_sales.values, marker='o',label='Ship Date')
plt.title('Monthly Sales Trends(Order Date vs Ship Date)')
plt.xlabel(' Month')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```

## Best Selling Products

```python
best_selling_products = sales_data.groupby(['Product Name','Product ID'])['Quantity'].sum().nlargest(10)
print("Top 10 Best-Selling Products:")
print(best_selling_products)
```

## Visualizing Best Selling Products

```python
plt.figure(figsize=(10, 6))
best_selling_products.plot(kind='bar', color='skyblue')
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Product Name, ID')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```
